```python
import numpy as np
import time
import matplotlib.pyplot as plt

def discrete_fourier_transform(x):
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    X = np.dot(e, x)
    return X

def benchmark_dft(x):
    start_time = time.time()
    discrete_fourier_transform(x)
    end_time = time.time()
    return end_time - start_time

def benchmark_fft(x):
    start_time = time.time()
    np.fft.fft(x)
    end_time = time.time()
    return end_time - start_time

# Input sizes for benchmarking
input_sizes = [2**i for i in range(2, 14)]  # Range from 4 to 32768

# Perform benchmarking for DFT and FFT
dft_times = []
fft_times = []
for N in input_sizes:
    x = np.random.rand(N)
    dft_time = benchmark_dft(x)
    fft_time = benchmark_fft(x)
    dft_times.append(dft_time)
    fft_times.append(fft_time)

# Plot the benchmark times
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, dft_times, label='DFT')
plt.plot(input_sizes, fft_times, label='FFT')
plt.xlabel('Input Size')
plt.ylabel('Time (seconds)')
plt.title('Benchmarking DFT and FFT')
plt.legend()
plt.xscale('log')  # Logarithmic scale for x-axis
plt.yscale('log')  # Logarithmic scale for y-axis
plt.show()
```
